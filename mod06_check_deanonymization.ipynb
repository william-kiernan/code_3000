# Discussion Questions

### Describe your reaction to the match rate. What sources of data might be prone to privacy issues?

The match rate of 0.421 shows that a large portion of the anonymized records could still be re-identified using only a few demographic variables. This demonstrates that removing names alone does not fully protect privacy because combinations of attributes such as age, gender, and ZIP code can uniquely identify individuals. Many real-world datasets could be vulnerable to this type of re-identification, especially datasets that contain demographic or location information.

### Provide examples of at least two other variables that could increase the match rate if included.

Additional variables would likely increase the match rate because they make each record more unique. For example, including full ZIP code instead of ZIP3 would make geographic identification more precise. Another variable that could increase the match rate is date of birth, which is often highly identifying when combined with other demographic attributes.
